```python
def new_policy(policy_index, file):
    one = []
    two = []
    three = []
    four = []
    five = []
    with open("test.csv","r") as file:
        csv_reader = csv.reader(file, delimiter = ',')
        for row in csv_reader:
            if len(row)!=0 and (row[0]==str(policy_index)):
                if(row[2][1] == '1'):
                    one.append(int(row[1]))
                elif(row[2][1] == '2'):
                    two.append(int(row[1]))
                elif(row[2][1] == '3'):
                    three.append(int(row[1]))
                elif(row[2][1] == '4'):
                    four.append(int(row[1]))
                elif(row[2][1] == '5'):
                    five.append(int(row[1]))
    return ([policy_index, {1: one, 2: two, 3: three, 4: four, 5:five}])
                
            
    
```

```python
def fetch_seg(lst, attr):
    name="policy_"+str(lst[0])+".yml"
    txt = ""
    with open('dataset/annotations/'+name) as file:
        segment_list = yaml.load(file,Loader=yaml.FullLoader)
        for d in segment_list['segments']:
            if d['segment_id'] in lst[1][attr]:
                txt += d['segment_text']
        with open('summary.csv', mode='a') as segment_file:
            segment_writer = csv.writer(segment_file)
            segment_writer.writerow([lst[0], attr, txt])
```

```python
policy_index_tracker = []
with open("test.csv","r") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter = ',')
    for row in csv_reader:
#         print(row[0])
#         print("\n\n")
        if len(row)!=0 and row[0] not in policy_index_tracker:
            f = (new_policy(row[0], csv_reader))
            for i in range(1,6):
                fetch_seg(f,i)
            policy_index_tracker.append(row[0])
        
```
