```python
import pandas as pd
import plotly.express as px
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!ls
```

```python
df = pd.read_csv('LOGS_CSV/loss/train_epoch_loss.csv')
df1=pd.read_csv('LOGS_CSV/loss/validation_epoch_loss.csv')
```

```python
df1['Step']
```

```python
# Plot with differently-colored markers.
plt.plot(df['Step'], df['Value'], 'b-', label='train')
plt.plot(df1['Step'],df1['Value'], 'g-', label='validate')

# Create legend.
plt.legend(loc='upper right')
plt.xlabel('Step')
plt.ylabel('Accuracy')
```

```python
# Plot with differently-colored markers.
sns.lineplot(df['Step'], df['Value'], label='train')
sns.lineplot(df1['Step'],df1['Value'], label='validate')

# Create legend.
plt.legend(loc='upper right')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.savefig('loss.jpeg')
```
